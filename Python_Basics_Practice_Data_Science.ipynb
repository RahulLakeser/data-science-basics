```python
# Question 49 — Find the Most Frequent Number

# Example:

# [10, 20, 10, 30, 20, 10]

# Counts:
# 10 -> 3
# 20 -> 2
# 30 -> 1

# Output:
# 10

# Rules:
# Use the same nested-loop counting pattern.
# No dictionaries.
# No count().
```

```python
lst = []
n = int(input("Enter the numbers you want to add in list : "))
for i in range(n):
  num = int(input(f"Enter number {i+1} : "))
  lst.append(num)


max_count = 0
for i in lst:
  count = 0
  for j in lst:
    if j == i:
     count+=1

  if count > max_count:
    max_count = count
    count_frequency = i


print(count_frequency)
```

```python
# Question 50 — Find the Least Frequent Number

# Example:
# [10, 20, 10, 30, 20, 40]

# Counts:
# 10 -> 2
# 20 -> 2
# 30 -> 1
# 40 -> 1

# Output:
# 30
```

```python
lst = []
n = int(input("Enter number how much numbers you want to add in list : "))
for i in range(n):
  num = int(input(f"Enter number {i+1} : "))
  lst.append(num)

least_freq = 0
min_count = len(lst)
for i in lst:

  count = 0

  for j in lst:
    if j == i:
      count+=1
  if count < min_count:
    min_count = count
    least_freq = i


print(f"Least frequency : {least_freq}")
```

```python
# Question 62 — Check Palindrome Using a Function

# Create a function:
# is_palindrome(word)

# It should:
# Return True if the word is a palindrome.
# Return False otherwise.
```

```python
def is_palindrome(word):
  reverse_word = ""
  for char in word:
    reverse_word = char + reverse_word

  if reverse_word == word:
    return True
  else :
    return False

result = is_palindrome("madam")

if result == True:
  print("Yes, this number is Palindrome")
else :
  print("No, this number is not Palindrome")

```

```python
# Question 63 — Count Even Numbers in a List Using a Function

# Create a function:
# count_even(numbers)

# It should:
# Take a list as a parameter.
# Return how many even numbers are in the list.
```

```python
def count_even(numbers):
  count = 0
  for i in numbers:
    if i % 2 == 0:
      count+=1
  return count
lst = [2,4,3,5,6,3,7]

result  = count_even(lst)

print(f"Total even numbers in list are : {result}")
```

```python
# 🎯 Question 64 — Return the Largest Number in a List

# Create a function:
# largest_in_list(numbers)

# It should:
# Take a list as a parameter.
# Return the largest number.
```

```python
def largest_in_list(numbers):
  n = numbers[0]
  for i in numbers:
    if i > n :
     n = i
  return n
lst = [9,23,40,38,87,57,37,25,48]
result = largest_in_list(lst)

print(f"The largest number of the list is  = {result}")
```

```python
# 🎯 Question 65 — Find the Second Largest Number in a List

# Create a function:
# second_largest(numbers)

# It should:
# Take a list as a parameter.
# Return the second largest number.
```

```python
def second_largest(numbers):
  first = numbers[0]
  second = numbers[0]
  for i in numbers:
    if i > first:
      first = i

    else:
      if i > second and second != first :
      second = i

  return second

lst = [39,13,3,14,8,12,18]

result = second_largest(lst)

print(result)

```
